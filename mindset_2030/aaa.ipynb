```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 2030      # Media
sigma = np.sqrt(8.36)  # Desviación estándar (raíz cuadrada de la varianza)

# Crear un rango de valores x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calcular la CDF de la distribución normal y reflejarla
y_cdf_reflejada = 1 - norm.cdf(x, mu, sigma)

# Graficar
plt.plot(x, y_cdf_reflejada, label=f'CDF Reflejada Normal($\mu$={mu}, $\sigma^2$={8.36})', color='green')
plt.title('Función de Distribución Acumulada Reflejada')
plt.xlabel('x')
plt.ylabel('Probabilidad acumulada')
plt.legend()
plt.grid(True)
plt.show()
```
