```python
import matplotlib.pyplot as plt


hs =  []
year = []
omega_point = 2040
human_live = 80
j = 1

for i in range(9,-5,-1):
    year.append(int(omega_point - 2**i*human_live))
    hs.append(j)
    j+=1
    
plt.figure(figsize=(18,10))   
plt.plot(hs,year, 'o-',  markerfacecolor = 'red', markersize = 12)
plt.xlabel('Puntos de Singularidad Histrica', fontsize=22)
plt.ylabel('Ao', fontsize=22)
plt.title('Punto Omega', fontsize=25)
plt.xticks(hs)
i = 0
while i < len(hs):
    if hs[i]< 5:
        Dhs = 0.2
        Dyear = -800
    else:
        Dhs = -0.2
        Dyear = -1800
    plt.annotate(year[i],(hs[i]+Dhs,year[i]+Dyear), fontsize=15)
    i += 1
plt.show()
```
